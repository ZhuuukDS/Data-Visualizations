```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.DataFrame({'vars': ['A', 'A', 'B', 'B'],
                   'quantity': [10, 12, 14, 16]})
```

```python
df
```

```python
# ðŸ¤ª
fig =plt.figure(figsize=(12,4))
plt.subplots_adjust(left=0.15)
fig.text(s='Barplots. So different...', x=0.4, y=1, fontsize=20)
plt.title('Let\'s assume we have this dataframe', size=16, y=0.75, loc='left')
plt.table(cellText=df.values,
                      rowLabels=df.index,
                      colColours=['gray', 'gray'],
                      rowColours=['#a1c9f4', '#a1c9f4', '#ffb482', '#ffb482'],
                      cellColours=[['#a1c9f4', '#a1c9f4'],['#a1c9f4', '#a1c9f4'], ['#ffb482', '#ffb482'],  ['#ffb482', '#ffb482']],
                      colLabels=df.columns,
                      loc='center',
         cellLoc='center')
plt.axis('off')

fig.text(x=0.15, y=0.25, s='And we want to plot a simple barplot', size=16)
fig.text(x=0.15, y=0, s='We can do it in different ways, and... ', fontsize=14)
fig.text(x=0.46, y=0, s=' with different results ', size=20, rotation=10,
         va="center",
         bbox=dict(boxstyle="round",
                   ec=(1., 0.5, 0.5),
                   fc=(1., 0.8, 0.8),
                   )
         )
plt.show()

print('\t\t\t\t\t\t\U0001F92A\n')

fig, axs = plt.subplots(2,2, figsize=(12,10))
#fig.suptitle('\n')
sns.barplot(data=df, x='vars', y='quantity', ci=None, ax=axs[0,0])
axs[0,0].set_title('Default Seaborn Barplot\n', size=14)
axs[0,0].set_xlabel('sns.barplot(data=df, x=\'vars\', y=\'quantity\')\nShows $mean$ value per category\n', size=12)
axs[0,0].xaxis.set_label_position('top')


sns.barplot(data=df, x='vars', y='quantity', ci=None, ax=axs[0,1], estimator=sum)
axs[0,1].set_title('Seaborn Barplot with $estimator=sum$ \n', size=14)
axs[0,1].set_xlabel('sns.barplot(data=df, x=\'vars\', y=\'quantity\', estimator=sum)\nShows $sum$ of values per category\n', size=12)
axs[0,1].xaxis.set_label_position('top')
                                  
df.plot.bar(x='vars', ax=axs[1,0], rot=0, legend=False)
axs[1,0].set_title('Pandas Barplot\n', size=14)
axs[1,0].set_xlabel('df.plot.bar()\nShows $all$ values per category\n', size=12)
axs[1,0].xaxis.set_label_position('top')

axs[1,1].bar('vars', 'quantity', data=df)
axs[1,1].set_title('Matplotlib Barplot\n', size=14)
axs[1,1].set_xlabel('plt.bar(x=\'vars\', height=\'quantity\', data=df)\nShows $max$ value per category\n', size=12)
axs[1,1].xaxis.set_label_position('top')

plt.subplots_adjust(wspace=0.2, hspace=0.6)

for ax in axs.flatten():
    ax.set_axisbelow(True)
    ax.set_ylabel(None)
    ax.grid()

plt.show()
```
