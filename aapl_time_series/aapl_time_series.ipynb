## Time Series Graph with Plotly
____

A time-series graph is a line plot used to visualize time series data. Time-series data is the data collected over an interval of time. Some of the popular examples of time series data are stock price data, daily sales data, daily covid cases data, etc.

While visualizing a time-series graph, the time intervals lie on the x-axis, and the data points collected over those time intervals lie on the y-axis. For example, while visualizing a time-series graph of stock price data, the date column will lie on the x-axis, and the price column will lie on the y-axis.

To visualize a time series graph using Python, I will be using a stock price dataset. There are many libraries in Python for data visualization; I will be using Plotly as it is easy to visualize interactive visualizations using plotly. So letâ€™s start this task by collecting the latest stock price data of Apple:

```python
import pandas as pd
import yfinance as yf #  API provided by Yahoo Finance to collect the latest stock price data
import datetime
from datetime import date, timedelta
import plotly.express as px

end_date = date.today() # today is the end day in our time series

start_date = date.today() - timedelta(days=360) # we'll start a time series from the day one year ago

data = yf.download('AAPL', 
                      start=start_date, 
                      end=end_date, 
                      progress=False)

data.index = data.index.strftime("%Y-%m-%d")

data.head()
```

The above data is collected by using the yfinance API. Now below is how we can visualize a time series graph using Python:

```python
figure = px.line(data, x = data.index, y = "Close")
#figure.write_image('time_series.png')
figure.show('png')
```
