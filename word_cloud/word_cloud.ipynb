## Word Cloud from a Pandas DataFrame
___

A word cloud is a data visualization technique that shows the most used words in large font and the least used words in small font. It helps to get an idea about the text data, especially when working on problems based on natural language processing. This notebook shows how to visualize a word cloud from a pandas DataFrame in Python.

A pandas DataFrame is used to store the data that we use when working on a data science task. Sometimes the dataset contains a column with textual information such as opinions or reviews of people about a product. To understand how most people think about the product, we can visualize a word cloud of that column. So to view a word cloud from a pandas DataFrame in Python, we need to have the `wordcloud` library installed in our Python environment. We can install this library using the pip command:

Now letâ€™s see how to visualize a word cloud from a pandas DataFrame in Python. For this task, I will first import all the necessary Python libraries and a dataset with textual information:

```python
from wordcloud import WordCloud
from wordcloud import ImageColorGenerator
from wordcloud import STOPWORDS
import matplotlib.pyplot as plt
import pandas as pd

data = pd.read_csv('https://raw.githubusercontent.com/amankharwal/Website-data/master/spam.csv')
data.head()
```

There are only two columns in this dataset where the text column contains textual data. So below is how we can visualize a word cloud from the text column of this dataset using Python:

```python
text = " ".join(i for i in data.text)
stopwords = set(STOPWORDS)

wordcloud = WordCloud(stopwords=stopwords, background_color="white").generate(text)

plt.figure(figsize=(10,8))

plt.imshow(wordcloud, interpolation='bilinear')
plt.axis("off")
#plt.savefig('wordcloud.png')
plt.show()
```

So this is how we can easily visualize a word cloud from any column of your dataset using Python.
