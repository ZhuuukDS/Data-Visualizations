## Choropleth Map Using Plotly
____

A choropleth map is a shaded map where the intensity of the colour indicates the intensity or quantity of a particular feature. In Data Science, the choropleth maps are used to visualize the distribution of a feature in a geographical area. 

Now let’s see how to visualize a choropleth map. For this task, I will be using the plotly library in Python. So let’s import the necessary libraries and the dataset that we need to visualize choropleth maps:

```python
import plotly.graph_objects as go
import pandas as pd

data = pd.read_csv('https://raw.githubusercontent.com/amankharwal/Website-data/master/usa.csv')

data.head()
```

The dataset used here is based on the agricultural exports of the United States. Now let’s see how to visualize choropleth maps with this dataset using Python:

```python
figure = go.Figure(data=go.Choropleth(locations=data["code"],
                                      z = data["total exports"].astype(float),
                                      locationmode="USA-states",
                                      colorscale="Reds",
                                      colorbar_title="Millions USD"))
figure.update_layout(title_text="US Agriculture Exports", geo_scope='usa')
#figure.write_image('choropleth.png')
figure.show('png')
```
