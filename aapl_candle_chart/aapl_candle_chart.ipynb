## Candlestick Chart
____

A candlestick chart is a data visualization tool used to analyze the price movements of stocks, cryptocurrencies, currencies, and other financial instruments.

As this chart is mainly used to analyze the price movements of financial instruments, so I will be using the Apple stock price data for this task. Let's take data on Appleâ€™s stock price for the past three months from Yahoo Finance.

```python
import pandas as pd
import yfinance as yf #  API provided by Yahoo Finance to collect the latest stock price data
import datetime
from datetime import date, timedelta
import plotly.graph_objects as go

end_date = date.today() # today is the end day in our time series

start_date = date.today() - timedelta(days=90) # we'll start a time series starting 30 days ago

data = yf.download('AAPL', 
                      start=start_date, 
                      end=end_date, 
                      progress=False)

data.index = data.index.strftime("%Y-%m-%d")
data['Date'] = data.index
data.reset_index(drop=True, inplace=True)
data.head()
```

```python
figure = go.Figure(data=[go.Candlestick(x=data["Date"],
                                        open=data["Open"], 
                                        high=data["High"],
                                        low=data["Low"], 
                                        close=data["Close"])])
figure.update_layout(title = "Apple Stock Price Analysis", 
                     xaxis_rangeslider_visible=False)
#figure.write_image('candlechart.png')
figure.show('png')
```

### Summary
So this is how we can use the `plotly` library in Python for visualizing a candlestick chart.
